## Importing Covid-19 data for the USA

#### The data was sourced from the following site https://datahub.io/core/covid-19

```python
import pandas as pd
```

### 1. Importing data for confirmed cases

```python
# csv containing confirmed cases

df = pd.read_csv("raw_data/us_confirmed.csv")

df.head()
```

```python
df.info()
```

```python
# Changing "Date" column to datetime

df["Date"] = pd.to_datetime(df["Date"])
df.info()
```

### 2. Importing data for deaths

```python
df2 = pd.read_csv("raw_data/us_deaths.csv")

df2
```

```python
df2.info()
```

```python
# Changing "Date" column to datetime

df2["Date"] = pd.to_datetime(df2["Date"])
df2.info()
```

### 3. Subsetting data into states

```python
df_confirmed = df[["Date", "Case", "Province/State"]]

df_confirmed = df_confirmed.loc[df["Date"] == "2020-07-04"]
df_confirmed = df_confirmed.groupby(["Province/State"]).agg({"Case":"sum"})
```

```python
df_confirmed = df_confirmed.reset_index()

df_confirmed.head()
```

```python
df_deaths = df2[["Date", "Case", "Province/State"]]
df_deaths = df_deaths.loc[df["Date"] == "2020-07-04"].groupby(["Province/State"]).agg({"Case":"sum"})
```

```python
df_deaths = df_deaths.reset_index()

df_deaths.head()
```

### 4. Combining data-sets

- first rename the column names so that they are unique

```python
df_confirmed.columns = ["State", "Confirmed_cases"]

df_confirmed.head()
```

```python
df_deaths.columns = ["State", "Deaths"]

df_deaths.head()
```

```python
df_covid_impact = df_confirmed.merge(df_deaths)

df_covid_impact
```

```python
df_covid_impact.info()
```

```python
df_covid_impact.to_csv("data/covid_us_impact_by_state.csv")
```
